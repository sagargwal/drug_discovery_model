```python
import pandas as pd 
from chembl_webresource_client.new_client import new_client
pd.set_option('display.max_row', None)
pd.set_option('display.max_column', None)
```

```python
target = new_client.target
target_query = target.search("coronavirus")
targets = pd.DataFrame.from_dict(target_query)
print(targets)
selected_target = targets.target_chembl_id[6]
# selected_target
```

```python
target_query


```

```python
activity = new_client.activity

res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
res
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.to_csv("coronavirus.csv",index=False)
df.shape
```

```python
df2 = df[df.standard_value.notna()]
```

```python
df2 = df2[df2.canonical_smiles.notna()]
```

```python
len(df2.canonical_smiles)
```

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr = df2_nr.reset_index(drop=True)
df2_nr

# df2_nr = df2[~df2['canonical_smiles'].duplicated()]
```

```python
len(df2_nr.canonical_smiles)
```

```python
df2_nr
```

```python
# df3.to_csv('acetylcholinesterase_02_bioactivity_data_preprocessed.csv', index=False)
```

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
# df3.to_excel("clear.xlsx")
df3
```

```python
bioactivity_class = []
# print(len(df2_nr.standard_value))
for i in df2_nr.standard_value:
    if float(i)>= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermidiate")
    
```

```python
bioactivity_class = pd.Series(bioactivity_class, name = "class")
len(bioactivity_class)
```

```python
df4 = pd.concat([df3, bioactivity_class],axis = 1)
# x = df4.loc[(df4["class"] == "")]

df4
```

```python
df4.to_csv("preprocessed_bioactivity4.csv",index = False)
```
